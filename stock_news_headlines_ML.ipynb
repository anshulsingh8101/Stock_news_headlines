## Model Implementation


(Random forest , XGBOOST and CATBoost)

```python
import pandas as pd
import numpy as np 
import warnings
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
import seaborn as sns
```

## XGboost model without any hpt

```python
max_features_num = [500,600,700,800,900,1000]
n_gram = [1,2,3,4,5]

start_time = time.time()

for i in max_features_num:
    for j in n_gram:
        countvectors_xg_no = CountVectorizer(ngram_range=(j,j), max_features=i)
        traindataset = countvectors_xg_no.fit_transform(headlines_train)
        testdataset = countvectors_xg_no.transform(headlines_test)

        model_xgb = XGBClassifier(random_state = 42)
        model_xgb.fit(pd.DataFrame(traindataset.todense() , columns = countvectors_xg_no.get_feature_names()), train['Label'])

        y_pred_xgb = model_xgb.predict(pd.DataFrame(testdataset.todense(), columns=countvectors_xg_no.get_feature_names()))
        score_xgb_no = accuracy_score(test['Label'], y_pred_xgb)
        matrix_xgb_no = confusion_matrix(test['Label'], y_pred_xgb)
        print('max number of features used : {}'.format(i))
        print('ngram range used : ({},{})'.format(j,j))
        print(score_xgb_no)
        print(matrix_xgb_no)
        

    
ttss = (time.time() - start_time)
print("Running Time: {}".format(datetime.timedelta(seconds=ttss)))
```

## XgBoost with hyper parameter tuning

```python
def performance(classifier, model_name):
    print(model_name)
    print('Best Score: ' + str(classifier.best_score_))
    print('Best Parameters: ' + str(classifier.best_params_))
```

```python
# XGboost and catboost require pd.DataFrame as input and not matrix
```

```python
countvectors_xg = CountVectorizer(ngram_range=(2,2), max_features = 500)
traindataset = countvectors_xg.fit_transform(headlines_train)
testdataset = countvectors_xg.transform(headlines_test)

start_time = time.time()

param_grid = { 'n_estimators': [500,550,600,650],
    'colsample_bytree': [0.75,0.8,0.85],
    'max_depth': [None,3, 5 , 10, 15],
    'reg_alpha': [1],
    'reg_lambda': [2, 5, 10],
    'subsample': [0.55, 0.6, .65,0.9],
    'learning_rate':[0.5],
    'gamma':[.5,1,2],
    'min_child_weight':[0.01],
    'sampling_method': ['uniform']}

xgboost_md = XGBClassifier(random_state = 42)
model_xg_cv = RandomizedSearchCV(estimator = xgboost_md, param_distributions = param_grid, cv=5, verbose = 2, n_jobs= 2)
model_xg_cv.fit(pd.DataFrame(traindataset.todense(),columns =countvectors_xg.get_feature_names()),  train['Label'])

xg_time = (time.time() - start_time)

print("Running Time: {}".format(datetime.timedelta(seconds=xg_time)))
performance(model_xg_cv,'XGB')
```

```python
model_xg_cv_best = model_xg_cv.best_estimator_
model_xg_cv_best.fit(traindataset,train['Label'])

y_pred_xgb = model_xg_cv_best.predict(testdataset)
```

```python
score_xgb = accuracy_score(test['Label'],y_pred_xgb )
matrix_xgb = confusion_matrix(test['Label'], y_pred_xgb)

print(score_xgb)
print(matrix_xgb)
```

```python
sns.heatmap(matrix_xgb, annot=True, fmt='g')
```

## Catboost Model 

```python
# the traindataset and the test data set will be the same as given for xgboost model
countvectors_cat = CountVectorizer(ngram_range=(2,2), max_features=500)
traindataset =countvectors_cat.fit_transform(headlines_train)
testdataset = countvectors_cat.transform(headlines_test)

start_time = time.time()

model_catboost = CatBoostClassifier(random_state = 42)
model_catboost.fit(pd.DataFrame(traindataset.todense(), columns = countvectors_cat.get_feature_names()), train['Label'])

catboost_time = (time.time() - start_time)

y_pred_catboost = model_catboost.predict(pd.DataFrame(testdataset.todense(), columns = countvectors_cat.get_feature_names()))
score_catboost_no = accuracy_score(test['Label'], y_pred_catboost)
matrix_catboost_no = confusion_matrix(test['Label'], y_pred_catboost)

print("Running Time: {}".format(datetime.timedelta(seconds=catboost_time)))
print(score_catboost_no)
```

## RandomForest with HPT.

```python
from sklearn.metrics import SCORERS
SCORERS.keys()
```

```python
start_time = time.time()

params = {'n_estimators': [100, 300, 400],
          'min_samples_split': [2,3,5],
          'min_samples_leaf': [1, 2, 5 ],
          'max_features': ['auto', 10],
          'max_depth': [None, 15, 20],
          'criterion': ['gini', 'entropy'],
          'bootstrap': [True,False]
         }

rf = RandomForestClassifier(random_state =42)
model_rf_cv = GridSearchCV(rf, param_grid = params, scoring = 'neg_mean_squared_error', cv = 5, verbose=2, n_jobs=1)
model_rf_cv.fit(traindataset, train['Label'])

rf_time = (time.time() - start_time)


performance(model_rf_cv, 'Random Forest')
```

```python
print("Running Time: {}".format(datetime.timedelta(seconds=rf_time)))
```
